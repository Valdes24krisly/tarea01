```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.plot([1,2,3,4,5,2,3,8])
plt.show()
```

```python
x=np.arage(10)
y=np.sin(x)
plt.plot(x,y,'ro-')
plt.xlabel('distancia(m)')
plt.ylabel('densidad(g/ml)')
plt.suplitle('experimento 1')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot([1,4,2,6])
plt.subplot(2,2,2)
plt.hist(np.random.randn(1000))
plt.subplot(2,2,3)
plt.hist(np.random.rand(10))
plt.subplot(2,2,4)
plt.plot([1,2,3])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot([1,4,2,6],'ro-')
plt.xlabel("x1")
plt.subplot(2,2,2)
plt.hist(np.random.randn(1000), color="orange")
plt.subplot(2,2,3)
plt.hist(np.random.rand(10))
plt.xlabel("x3(cm)")
plt.subplot(2,2,4)
plt.plot([1,2,3])
plt.suptitle("Datos Experimentales")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(np.random.rand(20), '-o', ms=20, lw=2, alpha=0.7, mfc='orange')
ax.grid()
fig.text(0.95, 0.05, 'Property of MPL',
         fontsize=50, color='gray',
         ha='right', va='bottom', alpha=0.5)
plt.show()
```
